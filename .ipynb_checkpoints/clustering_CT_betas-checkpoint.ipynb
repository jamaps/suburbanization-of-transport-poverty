```python
df = pd.read_csv("data/var_betas.csv")
df["pop_beta_1000s"] = ct["pop_beta"] / 1000
```

```python
dfcluster = df[["SESi_beta","Ai_combine_beta","dw_no_car_beta","mean_commute_time_adult_beta","R_activities_per_day_beta"]]
dfcluster = dfcluster.dropna()
```

```python
k = 2
cluster_metrics = []
while k <= 7:
    X = sklearn.preprocessing.StandardScaler().fit_transform(dfcluster)
    kmeans = sklearn.cluster.KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    sil_score_kmeans = sklearn.metrics.silhouette_score(X, kmeans.labels_, metric='euclidean')
    var_crit = sklearn.metrics.calinski_harabasz_score(X, kmeans.labels_)
    cluster_metrics.append([k,sil_score_kmeans,var_crit])
    k += 1
cluster_metrics = pd.DataFrame(cluster_metrics)
cluster_metrics.columns = ["k","sil_score","var_crit"]
```

```python
sns.relplot(x="k", y="var_crit", kind="line", data=cluster_metrics)
```
